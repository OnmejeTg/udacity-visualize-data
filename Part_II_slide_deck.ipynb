## Investigation Overview

> In this exploration I want to demostrate that there is a strong negative corrolation between
the BorrowersAPR and the Proser Score ie. the higher the prosper score the lower 
the APR. Also that the higher the loan amount the lower the borrowersAPR showing a negative 
corrolation between the two

## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. the loans were given by proser loan company

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_df = pd.read_csv('prosperLoanData.csv')
```

```python
# Extracting desired columns and droping na values
cols = ['Term', 'EmploymentStatus', 
        'DebtToIncomeRatio', 'StatedMonthlyIncome',
       'LoanOriginalAmount','ProsperScore', 'BorrowerAPR',]
loan_sub = loan_df[cols].dropna()
```

## Visualization 1- Destibution of Borrowers APR
> The Borrowers APR seem to be multimodal with a high spike around 0.37. The values typically ranges from o.05-0.43 which is a percentage of the annual rate charged

```python
# Ploting a histogram of borrowers APR
bins = np.arange(0, loan_sub.BorrowerAPR.max()+0.05, 0.01)
plt.figure(figsize=[8, 5])
plt.hist(data = loan_sub, x = 'BorrowerAPR', bins = bins);
plt.xlabel('Borrower APR')
plt.title('Borrowers APR');
```

## (Visualization 2- Destribution of Prosper score)

> The Prospers score is a custom risk score build using historical prosper data. It ranges from 1-11 with 11 being the best or lowest risk socre. The most common scores are 4,6,and 8 and the least common is 1

```python
# plotting a bar chart for the Prosper score
plt.figure(figsize=(8,5))
default_color = sb.color_palette()[0]
sb.countplot(data=loan_sub, x='ProsperScore', color = default_color);
```

## (Visualization 3- Destribution of loan amount)

> The loan original amount is destributd between 10000- 35000. Most of the values fall in the lower half of the amount. There also is spike in multiples of 5000, indicating most people borrow in multiples of 5000

```python
# Ploting a hsitogram for loan amount
bins = np.arange(8000, loan_sub.LoanOriginalAmount.max()+800, 800)
plt.figure(figsize=[12, 6])
plt.hist(data = loan_sub, x = 'LoanOriginalAmount', bins=bins);
plt.xlabel('Original loan amount ($)') 
plt.title('Loan Amount');
```

##   Borrowers APR vs. Prosper score
> There is a negative corrolation between the prosper score and the APR. Borrowers with better scores have lesser APR

```python
plt.figure(figsize=(8,5))
sb.boxplot(x=loan_sub.ProsperScore, y=loan_sub.BorrowerAPR, color=sb.color_palette()[0])
plt.title('Borrowers APR vs. Prosper score');
```

## Borrowers APR vs. Loan Amount
> There is also a negative reationship between the loan amount and the APR. Hence The more the loan amount the lower the APR. this could indicate a strategy to encourage borrowers to borrow more.

```python
plt.figure(figsize = [8, 6])
sb.regplot(data = loan_sub, x = 'LoanOriginalAmount', y = 'BorrowerAPR', scatter_kws={'alpha':0.01})
plt.title('APR vs Loan Amount');
```
